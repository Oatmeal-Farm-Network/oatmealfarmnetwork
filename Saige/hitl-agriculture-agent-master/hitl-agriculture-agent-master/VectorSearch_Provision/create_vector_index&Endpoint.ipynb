```python
from google.cloud import aiplatform

aiplatform.init(
    project="animated-flare-421518",  # Project ID from https://console.cloud.google.com/home/dashboard
    location="us-central1"      # choose your Vertex region
)
```

```python
# Only run this if you want to create new index
# create MatchingEngine index

index = aiplatform.MatchingEngineIndex.create_tree_ah_index(
    display_name="sage_index_005",         # human-readable name for the index
    description="we are using text embedding 005 for this index",                     
    contents_delta_uri= None,  # point this to a GCS location containing your serialized vectors
    dimensions=1536, # embedding dimension of text-embedding-005
    distance_measure_type= "DOT_PRODUCT_DISTANCE",
    approximate_neighbors_count = 150,
    leaf_node_embedding_count=1000,
    leaf_nodes_to_search_percent=7
)

print(f"Index Created: {index.resource_name}")
```

```python
# Only run this if you want to create new endpoint
# create Endpoint 
endpoint = aiplatform.MatchingEngineIndexEndpoint.create(
    display_name = "sage-index-endpoint",
    description = "Endpoint for text-embedding-005 semantic memory",
    public_endpoint_enabled=True
)

print("Endpoint Created:", endpoint.resource_name)
```
